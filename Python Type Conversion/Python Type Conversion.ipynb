```python
#Q1.Convert an integer to a floating-point number. 
#ANSWER:
num = 10
float_num = float(num)
print(float_num)   # 10.0
```

```python
#Q2.Convert a float to an integer. 
#ANSWER:
x = 12.9
y = int(x)
print(y)   # 12
```

```python
#Q3.Convert an integer to a string. 
#ANSWER:
num = 100
s = str(num)
print(s)   # "100"
```

```python
#Q4.Convert a list to a tuple. 
#ANSWER:
lst = [1, 2, 3]
t = tuple(lst)
print(t)   # (1, 2, 3)
```

```python
#Q5.Convert a tuple to a list. 
#ANSWER:
tup = (4, 5, 6)
lst = list(tup)
print(lst)   # [4, 5, 6]

```

```python
#Q6.Convert a decimal number to binary. 
#ANSWER:
n = 10
b = bin(n)
print(b)   # 0b1010
```

```python
#Q7.Convert a non-zero number to boolean. 
#ANSWER:
n = 5
b = bool(n)
print(b)   # True

```
